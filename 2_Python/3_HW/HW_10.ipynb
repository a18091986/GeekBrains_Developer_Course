```python
import pandas as pd
import random
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

###### Ваша задача перевести его в one hot вид без get_dummies?

```python
data.loc[data['whoAmI'] == 'human', 'whoAmI_hum'] = 1
data.loc[data['whoAmI'] != 'human', 'whoAmI_hum'] = 0
data.loc[data['whoAmI'] != 'human', 'whoAmI_robot'] = 1
data.loc[data['whoAmI'] == 'human', 'whoAmI_robot'] = 0
data.whoAmI_hum = data.whoAmI_hum.astype(int)
data.whoAmI_robot = data.whoAmI_robot.astype(int)
data.head()
```
